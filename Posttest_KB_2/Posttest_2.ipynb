### Ketentuan Analisi Data Ke 3
Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. 

```python
total_records = df.shape[0]
total_attributes = len(df.columns)
attributes = list(df.columns.values)
total_attributes_number = len(df.select_dtypes('int').columns)

info = {
    'Jumlah record': total_records,
    'Jumlah attribute': total_attributes,
    'Nama attribute': attributes,
    'Jumlah attribute bertipe angka': total_attributes_number,
}

for key, value in info.items():
    print(key, ":", value)
```

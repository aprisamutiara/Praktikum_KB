```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
```

Membaca file csv melalui variabel sephora dan menampilkan 5 data teratasnya

```python
sephora = pd.read_csv('sephora_website_dataset.csv')
sephora
```

## Visualisasi 1 (Bar Plot) 

Pada data di bawah ini menunjukkan harga produk di sephora website berdasarkan rating, terlihat bahwa sebagian besar orang-orang menyukai produk dari sephora website. Hal ini dibuktikan dengan jumlah produk sebanyak 175000 memiliki rating 4.5, 150000 dengan rating 4. dan untuk rating di bawah 3 tidak mencapai 25000. 

```python
data= sephora.groupby('rating').sum()

sns.barplot(x=data.index,
            y=data['price'])

plt.xticks(rotation=45, ha='right')
plt.title('Harga Produk Berdasarkan Rating')
plt.xlabel('Rating')
plt.ylabel('Price')
plt.show()
```

## Visualisasi 2 (Pie Chart)
Setengah dari presentase ditempati oleh rating 4 dan 4.5, rating 4.5 sebanyak 38.47% dan rating 4 sebanyak 31.12%. Sisanya ditempati oleh rating 0, 1, 1.5, 2, 2.5, 3 dan 3.5. 

```python
plt.figure(figsize=(15, 20))
sephora.rating.value_counts().plot.pie(autopct='%.2f%%')
plt.title('Presentase Rating Produk ')
plt.show()
```

## Visualisasi 3 (Heatmap)

Visualisasi di bawah ini, menunjukkan bahwa korelasi antara price dan value price lebih tinggi dibandingkan dengan korelasi lainnya yang cenderung lemah bahkan ada yang bernilai negatif.

```python
total_corr = sephora.loc[:, 'rating':'value_price'].corr()

sns.heatmap(
    data= total_corr,
    annot=True,
    cmap = 'coolwarm',
)
```

## Visualisasi 4 (Histogram)

Dari histogram di bawah menunjukkan bahwa produk dengan range harga 20-30 memiliki frekuensi lebih dari 2000 (nilai yang sering muncul). 1750 berkisar pada range 30-40. Sedangkan, range 40-100 memiliki frekuensi dibawah 1000. 

```python
plt.hist(sephora['price'], range=(0,100))
plt.title('Jumlah Harga Produk')
plt.xlabel('harga')
plt.ylabel('Frekuensi')
plt.show()
```

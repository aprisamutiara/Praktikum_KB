## Data Cleaning

### Cleaning NULL
Pada dataset di bawah ini, terdapat 12 nilai null pada kolom "value_price" dan 1 pada kolom "limited_time_offer"

```python
print('Jumlah record yang memiliki nilai null:')
sephora_manipulation.isna().sum()
```

Nilai null pada kolom "value_price" dan kolom "limited_time_offer" akan digantikan oleh modus pada masing masing jumlah kolomnya.

```python
modus = SimpleImputer(strategy='most_frequent')
```

Kemudian gunakan fungsi fit_transform untuk mengubah nilai null pada dataset sephora_manipulation di kolom "value_price" dan  "limited_time_offer" berdasarkan modus.

```python
sephora_manipulation[['value_price', 'limited_time_offer']] = modus.fit_transform (sephora_manipulation[['value_price', 'limited_time_offer']])
```

Terlihat pada data di bawah ini sudah tidak terdapat nilai null lagi. 

```python
print('Jumlah null setelah menggunakan SimpleImputer:')
sephora_manipulation.isna().sum()
```

## Data Transformation

Untuk melakukan data transformation, harus mengetahui tipe data setiap atribut terlebih dahulu. Atribut dengan tipe data numerik dapat dilakukan data transformation. 

Pada data di bawah ini, "rating", "number_of_reviews", "price", "value_price" dapat dilakukan transformasi data.

```python
sephora_manipulation.dtypes
```

"id" dan "love" berupa numerik tapi tidak bisa transformasi data, karena id memiliki data unik dan love merupakan target. 

"online_only", "exclusive", "limited_edition", "limited_time_offer" juga tidak bisa karena dari awal sudah bernilai 0 dan 1.

### Standarisasi

```python
from sklearn.preprocessing import StandardScaler
```

Melakukan proses transform data dengan standarisasi sehingga atribut "rating", "number_of_reviews", "price", "value_price" memiliki nilai skala 0 sampai 1.

```python
ss = StandardScaler()
sephora_std = sephora_manipulation.copy()
sephora_std[['rating', 'number_of_reviews', 'price', 'value_price']] = ss.fit_transform(sephora_std[['rating', 'number_of_reviews', 'price', 'value_price']])
```

```python
print('Nilai data sebelum scaling : ')
print(sephora_manipulation[['rating', 'number_of_reviews', 'price', 'value_price']])

print('Nilai data setelah scaling : ')
print(sephora_std[['rating', 'number_of_reviews', 'price', 'value_price']])
```

### Normalisasi 

```python
from sklearn.preprocessing import MinMaxScaler
```

Melakukan proses transform data pada atribut "price" dengan normalisasi sehingga nilai min menjadi 0 dan max menjadi 1.
